Q1. What is the difference between a function and a method in Python?

Ans: In Python, functions are different from a method in terms of where they're defined and how they're called:
-  FUNCTION:

Definition: A function is defined using the def keyword outside of a class.

Standalone: It is independent and not associated with any object.

Call syntax: Called directly using its name (e.g., my_function()).

example:

    def greet(name):
    return f"Hello, {name}!"

    print(greet("daksha"))  # Output: Hello, daksha!

- Method:

Definition: A method is a function defined inside a class.

Associated with objects: It operates on an instance (object) of the class and usually takes self as its first parameter.

Call syntax: Called on an object using dot notation (e.g., obj.method()).

Example:

    class Greeter:
    def greet(self, name):
        return f"Hello, {name}!"

    g = Greeter()
    print(g.greet("hitesh"))  # Output: Hello, hitesh!

Q6.  Explain the concept of generators in Python and how they are defined.

Ans: A generator is a special type of iterator that allows you to generate values on the fly instead of storing them all in memory at once. This is especially useful for working with large data or infinite sequences.

Generators are defined as:

1. Using a function with yield

- Instead of return, you use yield to pause the function and return a value.

- The function becomes a generator.
   
    def count_up_to(n):
       i = 1
    while i <= n:
        yield i
        i += 1
2. Calling it returns a generator object:


    gen = count_up_to(3)
    print(next(gen))  # 1
    print(next(gen))  # 2
    print(next(gen))  # 3


Q7. What are the advantages of using generators over regular functions?

Ans:

 1. Memory Efficiency

Generators use lazy evaluation, meaning they yield one item at a time and do not store the entire sequence in memory.

This is especially useful when working with large datasets or infinite sequences.

Example:

    def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1
This uses almost no memory compared to returning a full list.

2. Performance Improvement

Since generators yield values one at a time, they start producing results immediately, unlike regular functions that build and return the entire result set at once.

3. Infinite Sequences

Generators can produce infinite streams of data (like Fibonacci numbers or real-time logs) because they don’t compute everything at once.

4. Clean Syntax with yield

The yield keyword simplifies code that would otherwise require managing state manually using iterators and __iter__() / __next__() methods.

5. Pipeline Behavior (Chaining)
Generators can be easily chained together to form data pipelines, which is efficient and clean for data processing.

Example:

    squares = (x*x for x in range(100))     # Generator expression
6. Avoiding Temporary Storage
Regular functions often create temporary lists or structures that are discarded after use. Generators avoid this waste by yielding data on the fly.

Q9. Explain the purpose and usage of the `map()` function in Python.

Ans: The map() function in Python is used to apply a function to each item in an iterable (like a list, tuple, etc.) and returns a map object (an iterator) with the results.

Purpose:
To transform elements of an iterable by applying a function to each element, without using a for loop.

Syntax:

map(function, iterable)

- function: A function to apply (can be built-in, user-defined, or a lambda).

- iterable: A sequence (like list, tuple, etc.)

Example with Built-in Function:

    numbers = [1, 2, 3, 4]
    result = map(str, numbers)
    print(list(result))  # ['1', '2', '3', '4']
Example with Lambda Function:

    nums = [1, 2, 3, 4]
    squares = map(lambda x: x ** 2, nums)
    print(list(squares))  # [1, 4, 9, 16]
Example with Multiple Iterables:

    a = [1, 2, 3]
    b = [4, 5, 6]
    result = map(lambda x, y: x + y, a, b)
    print(list(result))  # [5, 7, 9]
Using map() vs. for loop:

-Using map

    result = list(map(lambda x: x * 2, [1, 2, 3]))  # [2, 4, 6]

-Equivalent with for loop

    result = []
    for x in [1, 2, 3]:
    result.append(x * 2)


Q10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

Ans: The map(), reduce(), and filter() functions in Python are all part of functional programming tools that work on iterables, but each serves a different purpose.

1. map() — Transform

Purpose:
Applies a function to each item in an iterable and returns a new iterable with the results.

Example:
    nums = [1, 2, 3]
    squares = list(map(lambda x: x ** 2, nums))
    print(squares)  # [1, 4, 9]

2. filter() — Select

Purpose:
Filters elements from an iterable based on a Boolean condition and returns only the items that satisfy the condition.

 Example:

    nums = [1, 2, 3, 4]
    evens = list(filter(lambda x: x % 2 == 0, nums))
    print(evens)  # [2, 4]

3. reduce() — Combine

Purpose:
Reduces an iterable to a single value by repeatedly applying a function that takes two arguments.

Example:

    from functools import reduce

    nums = [1, 2, 3, 4]
    total = reduce(lambda x, y: x + y, nums)
    print(total)  # 10

```python

def sum_of_even_numbers(numbers):
  return sum(num for num in numbers if num % 2 == 0)


```

Q4. Write a Python function that checks if a given number is prime or not from 1 to 200

```python
def is_prime(n):
    if n < 2 or n > 200:
        return False  # Only check within the range 1 to 200
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True
for num in range(1, 201):
    if is_prime(num):
        print(num, "is a prime number")
```

Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b
fib = FibonacciIterator(10)
for num in fib:
    print(num)
```

Q6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i
for value in powers_of_two(5):
    print(value)
```

Q9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit

```python
# List of temperatures in Celsius
celsius_temps = [0, 20, 30, 40, 100]

# Convert to Fahrenheit using map and lambda
fahrenheit_temps = list(map(lambda c: (c * 9/5) + 32, celsius_temps))

print("Fahrenheit temperatures:", fahrenheit_temps)
```

Q10. Create a Python program that uses `filter()` to remove all the vowels from a given string

```python
def remove_vowels(input_str):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda char: char not in vowels, input_str))

# Example usage
text = "Hello World"
result = remove_vowels(text)
print("String without vowels:", result)
```
